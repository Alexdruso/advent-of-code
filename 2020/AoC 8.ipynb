```python
f = open("input8.txt", "r")

program = f.read()
```

```python
program = [[line.split()[0], int(line.split()[1])] for line in program.splitlines()]

program
```

```python
def runUntilLoop(program):
    
    visited = [False]*len(program)
    index = 0
    acc = 0

    while index < len(program) and visited[index] == False:
        visited[index] = True
        if program[index][0] == 'acc':
            acc = acc + program[index][1]
            index = index+1
        elif program[index][0] == 'jmp':
            index = index +program[index][1]
        else: 
            index = index+1
    
    return index, acc
```

```python
runUntilLoop(program)
```

```python
acc = 0

for index in range(len(program)):
    if program[index][0] == 'jmp':
        program[index][0] = 'nop'
        stop, acc = runUntilLoop(program)
        if stop == len(program):
            break
        else: program[index][0] = 'jmp'
    elif program[index][0] == 'nop':
        program[index][0] = 'jmp'
        stop, acc = runUntilLoop(program)
        if stop == len(program):
            break
        else: program[index][0] = 'nop'
```
