```python
f = open("input15.txt", "r")
numbers =[int(number) for number in f.read().split(',')]

numbers
```

```python
lastAppearances={}

for turn in range(len(numbers)):
    lastAppearances[numbers[turn]] = (turn+1, turn+1)
    
lastAppearances
```

```python
for turn in range(len(numbers)+1, 2021):
    number = lastAppearances[numbers[turn-2]][0] - lastAppearances[numbers[turn-2]][1]
    numbers.append(number)
    lastAppearances.setdefault(number, (turn, turn))
    lastAppearances[number] = (turn, lastAppearances[number][0])
```
